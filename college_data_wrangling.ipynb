```python
path_college = '../../capstone_project_1/college-passing-statistics'
```

```python
%matplotlib inline
import pandas as pd
import statistics
import numpy as np
import os
import glob
```

```python
college_passing = pd.concat(map(pd.read_csv, glob.glob(path_college + '/*.csv')),sort = True)
```

```python
#split player names out
player_names = college_passing['Player']
player_names = player_names.iloc[:].str.split('*', n=1, expand=True)
player_names = player_names.iloc[:, 0].str.split('\\', n=1, expand = True)
player_names = player_names.drop([1], axis=1)
player_names = player_names.rename(columns={0:'Player'})
```

```python
college_passing['Player'] = player_names['Player']
```

```python
college_passing = college_passing.set_index(['Player']).sort_index()
```

```python
college_passing
```

```python
#total td
qb_td = college_passing['TD']
qb_td = qb_td.groupby(qb_td.index).sum()
#total int
qb_int = college_passing['Int']
qb_int = qb_int.groupby(qb_int.index).sum()
#total yds
qb_yds = college_passing['Yds']
qb_yds = qb_yds.groupby(qb_yds.index).sum()
#total games
qb_games = college_passing['G']
qb_games = qb_games.groupby(qb_games.index).sum()
#total att
qb_att = college_passing['Att']
qb_att = qb_att.groupby(qb_att.index).sum()
#total cmp
qb_cmp = college_passing['Cmp']
qb_cmp = qb_cmp.groupby(qb_cmp.index).sum()

#rush Att
rush = college_passing['rush_Att']
rush = rush.groupby(rush.index).sum()

#rush TD
td = college_passing['rush_TD']
td = td.groupby(td.index).sum()

#rush yds
rush_yds = college_passing['rush_Yds']
rush_yds = rush_yds.groupby(rush_yds.index).sum()
```

```python
#qb stats, replace necessary columns with totals
career_stats = college_passing.groupby(college_passing.index).mean().round(1)
career_stats['TD'] = qb_td
career_stats['Int'] = qb_int
career_stats['Yds'] = qb_yds
career_stats['G'] = qb_games
career_stats['Att']  = qb_att
career_stats['Cmp'] = qb_cmp
career_stats['rush_Att'] = rush
career_stats['rush_TD'] = td
career_stats['rush_Yds'] = rush_yds
```

```python
career_stats = career_stats.drop(['Rk'], axis=1)
```

```python
career_stats.to_csv(path_college+'college-passing.csv')
```

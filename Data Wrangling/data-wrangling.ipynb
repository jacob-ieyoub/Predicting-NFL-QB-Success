```python
path_college = '../../capstone_project_1/college-passing-statistics/'
path_data = '../../capstone_project_1/data/'
path_nfl = '../../capstone_project_1/nfl-passing-statistics/'
```

```python
%matplotlib inline
import pandas as pd
import statistics
import numpy as np
import os
import glob
```

# Create Testing Data

### College

```python
college_passing = pd.concat(map(pd.read_csv, glob.glob(path_college + 'college*.csv')),sort = True)
```

```python
#split player names out
player_names = college_passing['Player']
player_names = player_names.iloc[:].str.split('*', n=1, expand=True)
player_names = player_names.iloc[:, 0].str.split('\\', n=1, expand = True)
player_names = player_names.drop([1], axis=1)
player_names = player_names.rename(columns={0:'Player'})
```

```python
college_passing['Player'] = player_names['Player']
```

```python
college_passing = college_passing.set_index(['Player']).sort_index()
```

```python
college_passing
```

```python
#total td
qb_td = college_passing['TD']
qb_td = qb_td.groupby(qb_td.index).sum()
#total int
qb_int = college_passing['Int']
qb_int = qb_int.groupby(qb_int.index).sum()
#total yds
qb_yds = college_passing['Yds']
qb_yds = qb_yds.groupby(qb_yds.index).sum()
#total games
qb_games = college_passing['G']
qb_games = qb_games.groupby(qb_games.index).sum()
#total att
qb_att = college_passing['Att']
qb_att = qb_att.groupby(qb_att.index).sum()
#total cmp
qb_cmp = college_passing['Cmp']
qb_cmp = qb_cmp.groupby(qb_cmp.index).sum()
```

```python
#qb stats, replace necessary columns with totals
career_stats = college_passing.groupby(college_passing.index).mean().round(1)
career_stats['TD'] = qb_td
career_stats['Int'] = qb_int
career_stats['Yds'] = qb_yds
career_stats['G'] = qb_games
career_stats['Att']  = qb_att
career_stats['Cmp'] = qb_cmp
```

```python
career_stats = career_stats.drop(['Rk'], axis=1)
```

```python
career_stats
```

```python
career_stats.to_csv(path_data+'college-1997-2016.csv')
```

### NFL

```python
nfl = pd.concat(map(pd.read_csv, glob.glob(path_nfl + 'pass*.csv')), sort=True)
```

```python
nfl = nfl[(nfl['Att'] >=100)]
```

```python
nfl = nfl.set_index(['Player']).sort_index()
nfl.index = nfl.index.str.split('+', n=1, expand=True)
nfl = nfl.droplevel(1)
```

```python
nfl.index = nfl.index.str.split('*', n=1, expand=True)
nfl = nfl.droplevel(1)
```

```python
nfl.index = nfl.index.str.split('\\', n=1, expand=True)
nfl = nfl.droplevel(1)
```

```python
nfl.index.name = 'Player'
```

```python
#calculate career win percentage
qb_record = nfl['QBrec']
qb_record = qb_record.iloc[:].str.split(r'\-|/', n = 1, expand = True)
qb_record[1] = qb_record.iloc[:,1].str.split(r'\-|/', n=1, expand = True)
qb_record = qb_record.rename(columns={0: 'win', 1:'loss'})
qb_record[['win', 'loss']] = qb_record[['win','loss']].astype(int)
qb_record = qb_record.groupby(level=0).sum()
qb_record['win_percentage'] = (qb_record['win'] / (qb_record['win'] + qb_record['loss']))
```

```python
#qb total int 
qb_int = nfl['Int']
qb_int = qb_int.groupby(qb_int.index).sum()

#qb total td
qb_td = nfl['TD']
qb_td = qb_td.groupby(qb_td.index).sum()

#qb total yards
qb_yards = nfl['Yds']
qb_yards = qb_yards.groupby(qb_yards.index).sum()

#qb total games
qb_games = nfl['G']
qb_games = qb_games.groupby(qb_games.index).sum()

#qb sacks
qb_sacks = nfl['Sk']
qb_sacks = qb_sacks.groupby(qb_sacks.index).sum()
```

```python
#nfl average passer rating
rate_avg = nfl['Rate'].mean()

#nfl average ypa
ypa_avg = nfl['Y/A'].mean()

#nfl average aya
aya_avg = nfl['AY/A'].mean()

```

```python
ypa_avg = round(ypa_avg, 2)
rate_avg = round(rate_avg, 2)

print(ypa_avg)
print(rate_avg)
```

```python
#qb average for qbs playing from 2009-2019, replace necessary columns with career totals 
nfl = nfl.groupby(level=0).mean().round(1)
nfl.sort_values(by=['Y/A'], ascending=False)
nfl['win_pct'] = qb_record['win_percentage']
nfl['Int'] = qb_int
nfl['TD'] = qb_td
nfl['Yds'] = qb_yards
nfl['G'] = qb_games
nfl['Sk'] = qb_sacks
```

```python
nfl.head()
```

```python
nfl.to_csv('nfl-passing-1998-2017')
```

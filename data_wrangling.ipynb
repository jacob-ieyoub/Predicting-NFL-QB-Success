```python
path = '../../capstone_project_1/data'
```

```python
%matplotlib inline
import pandas as pd
import statistics
import os
import glob
import numpy as np
```

```python
nfl = pd.read_csv(path + '/nfl-passing.csv')
college = pd.read_csv(path + '/college-passing.csv')
```

```python
#rename column
nfl = nfl.rename(columns={'Unnamed: 0':'Player'})
#fill only missing value with average of column
nfl['QBR'] = nfl['QBR'].fillna(nfl['QBR'].mean())
```

```python
#merge dataframes, players in both
df = pd.merge(nfl, college, on='Player', how='inner', suffixes = ('_NFL','_College'))
```
